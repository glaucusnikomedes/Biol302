```python
def top_n(dd, n=10):
    '''Return a list of up to n (key, value) pairs from dictionary dd
       ordered by highest value.
    '''
    dd_sort = sorted(dd.items(), key=lambda x: (-x[1], x[0]))
    return dd_sort[:n]
```
